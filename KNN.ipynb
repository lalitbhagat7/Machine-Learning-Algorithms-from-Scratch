```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,qry_pnt,k):
    
    vals=[]
    
    m = X.shape[0]
    
    for i in range(m):
        vals.append((dist(qry_pnt,X[i]),Y[i]))
        
    vals.sort()
    
    vals=vals[:k]
    
    vals=np.array(vals)
    
    new_vals=np.unique(vals[:,1],return_counts=True)
    
    index=new_vals[1].argmax()
    
    pred=new_vals[0][index]
    
    return pred



    
    
    
    
```

```python
knn(X,Y,X_query,5)
```
